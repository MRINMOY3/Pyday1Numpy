```python
rads=np.linspace(0,2*np.pi,60)
print(rads)
```

```python
angles=np.rad2deg(rads)
print(angles)

plt.plot(angles,angles**2)
```

```python
radians=np.linspace(0,2*np.pi,60)
print(radians)
deg=np.rad2deg(radians)
print(deg)
```

```python
sin=np.sin(radians)
plt.plot(angles,sin,label='Sine')
plt.legend()
plt.show()
```

```python
deg=np.arange(0,361,15)
print(deg)
```

```python
rad=np.deg2rad(deg)
```

```python
print(rad)
```

```python
sin=np.sin(rad)
cos=np.cos(rad)
```

```python
plt.plot(rad,sin,label='Sine')
plt.legend()
plt.show()
```

```python
plt.plot(rad,cos,label='Sine')
plt.legend()
plt.show()
```

```python
plt.figure(dpi=100)
plt.plot(rad,sin,'b--o',label='Sine')
plt.plot(rad,cos,'r--^',label='Sine')
plt.legend()
plt.xticks(deg,rotation=90)
plt.plot()
```
